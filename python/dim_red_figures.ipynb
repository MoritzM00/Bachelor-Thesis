```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

n = 100

phi = np.linspace(0, 2 * np.pi, n)
theta = np.linspace(0, 2 * np.pi, n)
phi, theta = np.meshgrid(theta, phi)
R, r = 2, 1
x = np.cos(theta) * (R + r * np.cos(phi))
y = np.sin(theta) * (R + r * np.cos(phi))
z = r * np.sin(phi)

fig, (ax1, ax2)= plt.subplots(1, 2, figsize=(5.91, 4), subplot_kw=dict(projection='3d'))
lim = (max(abs(max(np.max(x), np.max(y), np.max(z))), abs(min(np.min(x), np.min(y), np.min(z)))))
ax1.plot_surface(x, y, z, rstride=5, cstride=5, cmap="viridis", antialiased=True)
ax1.view_init(36, 26)
ax1.set_zlim(-lim, lim)
ax1.axis("off")

ax2.plot_surface(x, y, z, rstride=5, cstride=5, cmap="viridis", antialiased=True)
ax2.view_init(-30, 45, 0)
ax2.set_zlim(-lim, lim)#
ax2.axis("off")
fig.savefig("../figures/torus.pdf", bbox_inches="tight", dpi=300)
plt.show()
```
