```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(["science", "grid", "vibrant"])

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Rectified Linear Units (ReLU)
def relu(x):
    return np.maximum(0, x)

# Tangens Hyperbolicus (tanh)
def tanh(x):
    return np.tanh(x)

# Plotting the functions
x = np.linspace(-10, 10, 100)


fig, axs = plt.subplots(1, 3, figsize=(5.91, 2), sharex=True)

axs[0].plot(x, sigmoid(x))
axs[0].text(0.5, -0.35, "a) Sigmoid", fontsize=11, ha="center", transform=axs[0].transAxes)

axs[1].plot(x, tanh(x))
axs[1].text(0.5, -0.35, "c) tanh", fontsize=11, ha="center", transform=axs[1].transAxes)

axs[2].plot(x, relu(x))
axs[2].text(0.5, -0.35, "b) ReLU", fontsize=11, ha="center", transform=axs[2].transAxes)
plt.tight_layout()
plt.savefig("../figures/activations.pdf", bbox_inches="tight")
plt.show()
```
